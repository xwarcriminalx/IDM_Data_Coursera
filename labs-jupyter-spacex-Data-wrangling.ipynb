# Lab 3 - Data Wrangling

Create the training label `Class` (1 = successful landing, 0 = unsuccessful).

```python
import pandas as pd
import numpy as np
df = pd.read_csv('https://cf-courses-data.s3.us.cloud-object-storage.appdomain.cloud/IBM-DS0321EN-SkillsNetwork/datasets/dataset_part_1.csv')
df.head()
```

### Number of launches per site

```python
df['LaunchSite'].value_counts()
```

### Number and occurrence of each orbit

```python
df['Orbit'].value_counts()
```

### Number and occurrence of each landing outcome

```python
landing_outcomes = df['Outcome'].value_counts()
landing_outcomes
```

### Set of unsuccessful (bad) outcomes

```python
bad_outcomes = set(landing_outcomes.keys()[[1,3,5,6,7]])
bad_outcomes
```

### Create the landing class label

```python
landing_class = [0 if outcome in bad_outcomes else 1 for outcome in df['Outcome']]
df['Class'] = landing_class
print('success rate:', df['Class'].mean())
df[['FlightNumber','Orbit','LaunchSite','Outcome','Class']].head()
```

```python
df.to_csv('dataset_part_2.csv', index=False)
print('saved dataset_part_2.csv')
```
