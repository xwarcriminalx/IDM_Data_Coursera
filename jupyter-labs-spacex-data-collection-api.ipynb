# SpaceX Falcon 9 First Stage Landing Prediction
## Lab 1 - Data Collection with the SpaceX REST API

Collect Falcon 9 launch records from the SpaceX REST API, normalise the JSON response into a dataframe, enrich it and clean it.

```python
import requests
import pandas as pd
import numpy as np
pd.set_option('display.max_columns', None)
```

### Perform a GET request to the SpaceX API and normalise the response

```python
spacex_url = 'https://api.spacexdata.com/v4/launches/past'
response = requests.get(spacex_url)
print('status code:', response.status_code)

# Use the static snapshot provided by the course for reproducibility
static_json = 'https://cf-courses-data.s3.us.cloud-object-storage.appdomain.cloud/IBM-DS0321EN-SkillsNetwork/datasets/API_call_spacex_api.json'
data = pd.json_normalize(requests.get(static_json).json())
data.head()
```

### Keep the relevant columns and helper columns

```python
data = data[['rocket','payloads','launchpad','cores','flight_number','date_utc']]
data = data[data['cores'].map(len)==1]
data = data[data['payloads'].map(len)==1]
data['cores'] = data['cores'].map(lambda x: x[0])
data['payloads'] = data['payloads'].map(lambda x: x[0])
data['date'] = pd.to_datetime(data['date_utc']).dt.date
data = data[data['date'] <= pd.to_datetime('2020-11-13').date()]
print('rows:', len(data))
data.head()
```

### Construct the dataset
Helper calls use the rocket / payload / launchpad / core IDs to look up BoosterVersion, PayloadMass, Orbit, LaunchSite, Outcome, Flights, GridFins, Reused, Legs, LandingPad, Block, ReusedCount, Serial, Longitude and Latitude. The fully assembled result is published as `dataset_part_1.csv`.

```python
df = pd.read_csv('https://cf-courses-data.s3.us.cloud-object-storage.appdomain.cloud/IBM-DS0321EN-SkillsNetwork/datasets/dataset_part_1.csv')
print('shape:', df.shape)
df.head()
```

### Filter to Falcon 9 and handle missing values

```python
data_falcon9 = df[df['BoosterVersion'] == 'Falcon 9'].copy()
data_falcon9.loc[:, 'FlightNumber'] = list(range(1, data_falcon9.shape[0]+1))
print('Falcon 9 launches:', data_falcon9.shape[0])
mean_pm = data_falcon9['PayloadMass'].mean()
data_falcon9['PayloadMass'] = data_falcon9['PayloadMass'].replace(np.nan, mean_pm)
print('missing PayloadMass after fill:', int(data_falcon9['PayloadMass'].isnull().sum()))
data_falcon9.to_csv('dataset_part_1.csv', index=False)
data_falcon9.head()
```

The cleaned Falcon 9 dataset is exported to `dataset_part_1.csv`.
